```python
import numpy as np
months = np.array(['JAN','FEB','MAR','MAY','APR','JUN','JUL','AUG','OCT','SEP','NOV','DEC'])
sales = []
```

```python
print("Enter sales (in $1000) in each months")
for month in months:
   value = float(input(f"{month}: "))
   sales.append(value)
```

```python
sale = np.array(sales)
print("total sale in year: ", np.sum(sale))
print("average of sale in year: ", np.mean(sale))
print("maximum sale of year:", np.max(sale))
print("minimum sale  of year:", np.min(sale))

best_month = months[np.argmax(sale)]
worst_month = months[np.argmin(sale)]
print("best month of year:" , best_month)
print("worst month of year: ", worst_month)

above_avg = months[sale > np.mean(sale)]
below_avg = months[sale < np.mean(sale)]
print("above average sales: ", above_avg)
print("below average sales: ", below_avg)
```
